### 행 단위 추출

* loc : 인덱스 기준으로 행 데이터 추출
* iloc : 행 번호 기준으로 행 데이터 추출

### loc : 인덱스로 행 추출

```python
### 인덱스와 행번호

df.head()
```

```python
df.loc[1] # 1번 인덱스
```

```python
# 마지막 인덱스에 해당하는 데이터
number_of_rows = df.shape[0]
df.loc[number_of_rows-1]
```

```python
df.tail(1)
```

```python
# 인덱스가 0, 99, 999인 데이터를 한번에 추출하려면
df.loc[[0,99,999]]
type(df.loc[[0,99,999]]) # 여러 줄 출력하면 df
```

tail method와 loc속성이 반환하는 자료형은 다르다.
* tail이 반환하는 데이터 자료형은 데이터 프레임
* loc이 반환하는 데이터 자료형은 시리즈

```python
subset_loc = df.loc[0]

subset_tail = df.tail(1)

print(type(subset_loc))
print(type(subset_tail))
```

### iloc : 행 번호로 행 데이터 추출

```python
print(df.iloc[-1]) # 행 순서이기 때문에 -1을 하면 가장 마지막 값을 가져옴  loc은 -1 인덱스가 없을 경우 error
```

```python
df.iloc[[0,99,999]]
```

### df.loc[[행],[열]] or df.iloc[[행], [열]]

```python
subset = df.loc[:,['year','pop','continent']]
subset.head()
```

```python
subset = df.iloc[[0,3,5],[2,4,1]] # column도 column명을 넣는게 아니라 column의 해당 순서를 넣어줘야 함
subset.head()
```
